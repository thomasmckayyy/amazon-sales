```python
# Imports

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import CSV file

sales = pd.read_csv('../Amazon Sales/amazon.csv')
sales.head(5)
```

```python
#Describing data

sales.columns
sales.dtypes
sales.describe()
```

```python
sales.columns
df = sales[['product_id', 'product_name', 'category', 'discounted_price',
       'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()
df.head(5)
```

```python
#rename column names
df.rename(columns={'product_id':'id',
                   'product_name':'item',
                  'discounted_price':'discount',
                  'actual_price':'price',
                  'rating_count':'r_count'})
```

```python
# Finding Sum of Null values.
df.isna().sum()
```

```python
# Finding Null values
df.isna()
```

```python
#Finding duplicated under product ID
df.loc[df.duplicated(subset=['product_id'])].head(5)
```

```python
# Finding where B07JW9H4J1 is duplicated and number of instances.

df.query('product_id == "B07JW9H4J1"')
```

```python
# Located duplicates, take them out, reset index so it matches number of rows

df = df.loc[~df.duplicated(subset=['product_id','rating_count'])] \
    .reset_index(drop=True).copy()
df
```
